Data Preparation adalah tahapan untuk membersihkan dan menyiapkan data sebelum dilakukan proses analisis atau pemodelan. 6.1 Handling Missing Values

```python
for col in df.select_dtypes(include=['float64','int64']):
    df[col] = df[col].fillna(df[col].median())

for col in df.select_dtypes(include=['object']):
    df[col] = df[col].fillna(df[col].mode()[0])

df.isna().sum()
```

Kode di atas digunakan untuk menangani missing values pada dataset. Untuk kolom numerik (float64 dan int64), nilai kosong diisi menggunakan median karena median lebih stabil terhadap outlier. Sedangkan untuk kolom kategorikal (object), nilai kosong diisi menggunakan modus atau nilai yang paling sering muncul.

Setelah proses imputasi dilakukan, pengecekan ulang menunjukkan bahwa seluruh kolom sudah tidak memiliki missing values, sehingga dataset siap untuk tahap selanjutnya.

6.2 Menghapus Duplikat

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

Kode ini digunakan untuk menghapus baris data yang terduplikasi agar tidak mempengaruhi hasil analisis. Setelah dilakukan pengecekan ulang, tidak ditemukan data duplikat, sehingga dataset tetap dalam kondisi konsisten.

6.3 Menangani Outlier (Metode IQR)

```python
for col in df.select_dtypes(include=['float64','int64']):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower, upper)
```

Kode ini digunakan untuk menangani outlier menggunakan metode Interquartile Range (IQR). Nilai yang berada di luar batas bawah dan batas atas (berdasarkan perhitungan IQR) tidak dihapus, tetapi dibatasi (clipping) agar tetap berada dalam rentang yang wajar.

Metode ini dipilih agar distribusi data tidak terlalu dipengaruhi oleh nilai ekstrem, namun tetap mempertahankan jumlah data yang ada.

```python
features = df.select_dtypes(include=['float64','int64'])
features.head()
```

Pada tahap ini dilakukan pemilihan variabel numerik yang akan digunakan untuk proses clustering. Algoritma K-Means hanya dapat memproses data numerik, sehingga kolom bertipe numerik dipilih sebagai fitur analisis.

6.5 Standardisasi Data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

Standarisasi dilakukan menggunakan StandardScaler() untuk menyamakan skala antar variabel numerik. Hal ini penting karena algoritma K-Means sensitif terhadap perbedaan skala. Jika tidak dilakukan standarisasi, variabel dengan nilai besar seperti price dapat mendominasi hasil clustering.

Dengan standarisasi, setiap variabel memiliki rata-rata 0 dan standar deviasi 1, sehingga proses clustering dapat berjalan lebih optimal dan menghasilkan segmentasi yang lebih akurat.

**Menentukan Jumlah Cluster (Elbow Method)**

Menentukan jumlah cluster merupakan langkah penting sebelum menerapkan algoritma K-Means. Pemilihan jumlah cluster yang tepat akan mempengaruhi kualitas segmentasi yang dihasilkan. Salah satu metode yang umum digunakan untuk menentukan jumlah cluster optimal adalah Elbow Method.

7.1 Pilih Fitur Numerik untuk Clustering

```python
features = df[['quantity', 'price', 'review_score', 'age']]
features.head()
```

Pada tahap ini dipilih variabel numerik yang relevan untuk proses clustering, yaitu quantity, price, review_score, dan age. Pemilihan fitur ini dilakukan karena variabel tersebut dianggap dapat merepresentasikan pola pembelian dan karakteristik pelanggan.

Algoritma K-Means hanya dapat memproses data numerik, sehingga kolom kategorikal tidak digunakan dalam tahap ini. Output features.head() ditampilkan untuk memastikan bahwa fitur yang dipilih sudah sesuai.

7.2 Standardisasi Data

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

Standarisasi dilakukan untuk menyamakan skala antar variabel. Hal ini penting karena variabel seperti price memiliki rentang nilai yang jauh lebih besar dibandingkan variabel lain seperti quantity.

Dengan menggunakan StandardScaler(), setiap variabel diubah sehingga memiliki rata-rata 0 dan standar deviasi 1. Proses ini mencegah satu variabel mendominasi proses pembentukan cluster.

7.3 Elbow Method

```python
inertia = []

K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K_range, inertia, marker='o')
plt.xlabel('Jumlah Cluster (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

Elbow Method digunakan untuk menentukan jumlah cluster optimal dengan melihat nilai inertia pada berbagai jumlah cluster (K). Inertia merupakan ukuran seberapa rapat data dalam cluster, di mana nilai yang lebih kecil menunjukkan cluster yang lebih kompak.

Pada kode di atas, dilakukan perulangan dari K = 1 hingga K = 10. Untuk setiap nilai K, model K-Means dijalankan dan nilai inertia disimpan dalam list. Selanjutnya, nilai inertia diplot dalam bentuk grafik.

Berdasarkan grafik yang dihasilkan, terlihat bahwa penurunan inertia sangat tajam hingga sekitar K = 3 atau K = 4, kemudian mulai melandai. Titik di mana penurunan mulai melambat disebut sebagai “elbow point”. Dari grafik tersebut, jumlah cluster optimal yang dipilih adalah K = 3 karena setelah titik tersebut penurunan inertia tidak terlalu signifikan.

Dengan demikian, proses selanjutnya adalah menerapkan algoritma K-Means menggunakan jumlah cluster yang telah ditentukan.

**K-Means Clustering**

K-Means Clustering adalah metode unsupervised learning yang digunakan untuk mengelompokkan data ke dalam beberapa cluster berdasarkan kemiripan karakteristiknya. Algoritma ini bekerja dengan cara membagi data ke dalam sejumlah K cluster, di mana setiap data akan ditempatkan pada cluster dengan jarak terdekat terhadap centroid (titik pusat cluster).

Tujuan dari K-Means pada penelitian ini adalah untuk mengelompokkan pelanggan berdasarkan pola pembelian mereka sehingga dapat diketahui segmentasi pelanggan yang memiliki karakteristik serupa.

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

df['Cluster'] = clusters

df.head()
```

Pada kode di atas, algoritma K-Means diterapkan dengan jumlah cluster sebanyak 3, sesuai hasil dari Elbow Method yang telah dilakukan sebelumnya. Parameter random_state=42 digunakan agar hasil clustering konsisten setiap kali kode dijalankan.

Fungsi fit_predict() digunakan untuk:

1. Melatih model K-Means pada data yang telah distandarisasi (X_scaled).

2. Menghasilkan label cluster untuk setiap data.

Hasil clustering kemudian ditambahkan ke dalam dataset sebagai kolom baru bernama Cluster. Berdasarkan output yang ditampilkan, setiap baris data kini memiliki label cluster berupa angka 0, 1, atau 2.

Angka tersebut merepresentasikan kelompok pelanggan yang memiliki pola pembelian serupa berdasarkan variabel quantity, price, review_score, dan age.

Dengan terbentuknya cluster ini, dataset telah berhasil disegmentasikan dan siap untuk dianalisis lebih lanjut guna mengetahui karakteristik masing-masing kelompok pelanggan.

# **8. Cek Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64']).columns
cols = [col for col in cols if 'id' not in col.lower()]

for col in cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = round((len(outliers)/len(df))*100, 2)

    results.append({
        "Kolom": col,
        "Persentase Outliers (%)": percent_outliers
    })

results_df = pd.DataFrame(results).set_index('Kolom')
results_df
```

Insight:

Berdasarkan metode IQR pada variabel numerik (selain ID), tidak ditemukan nilai yang berada di luar batas bawah maupun batas atas.

Hal ini menunjukkan bahwa distribusi nilai transaksi relatif stabil dan tidak terdapat pembelian ekstrem yang dapat mendistorsi analisis.

Dengan kondisi ini, data dapat langsung digunakan untuk proses eksplorasi dan clustering tanpa perlu penanganan khusus terhadap nilai ekstrem.

# **DATA VISUALIZATION**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('synthetic_online_retail_data_bersih.csv')

df.head()
```

Dataset ini merupakan dataset yang sudah melalui proses data cleaning, construct data, dan data reduction pada checkpoint sebelumnya, sehingga data sudah siap digunakan untuk visualisasi.

Kemudian data yang digunakan untuk visualisasi adalah clustering_data yang berisi variabel:



*   total_quantity
*   total_revenue
*   avg_price
*   age








```python
clustering_data = df[['total_quantity','total_revenue','avg_price','age']]
```

# **1. Bar Chart**

**Melihat rata-rata nilai variabel pelanggan**

```python
avg_values = clustering_data.mean()

plt.figure(figsize=(8,5))
plt.bar(avg_values.index, avg_values.values)

plt.title('Rata-rata Variabel Pelanggan')
plt.xlabel('Variabel')
plt.ylabel('Nilai Rata-rata')

plt.show()
```

**Insight:**

Grafik menunjukkan bahwa rata-rata total_revenue memiliki nilai paling tinggi dibandingkan variabel lainnya. Hal ini menandakan bahwa kontribusi nilai transaksi pelanggan cukup besar dibandingkan jumlah barang yang dibeli maupun harga rata-rata produk.

**Action:**

Perusahaan dapat mempertahankan pelanggan dengan nilai revenue tinggi melalui program loyalitas, promo khusus, atau penawaran eksklusif agar pelanggan tetap melakukan pembelian.

# **2. Pie Chart**

**Melihat proporsi total revenue tiap kelompok usia**

```python
age_group = pd.cut(clustering_data['age'], bins=[18,25,35,45,60],
                   labels=['18-25','26-35','36-45','46-60'])

revenue_by_age = clustering_data.groupby(age_group, observed=False)['total_revenue'].sum()

plt.figure(figsize=(6,6))
plt.pie(revenue_by_age, labels=revenue_by_age.index, autopct='%1.1f%%')

plt.title('Proporsi Revenue Berdasarkan Kelompok Usia')

plt.show()
```

**Insight:**

Sebagian besar revenue berasal dari kelompok usia produktif (26–45 tahun). Hal ini menunjukkan bahwa kelompok usia tersebut memiliki aktivitas belanja yang lebih tinggi dibandingkan kelompok usia lainnya.

**Action:**

Perusahaan dapat memfokuskan strategi pemasaran pada kelompok usia produktif dengan promosi yang lebih relevan, seperti diskon produk populer atau rekomendasi produk sesuai kebutuhan mereka.

# **3. Line Chart**

**Melihat tren rata-rata revenue berdasarkan usia**

```python
age_revenue = clustering_data.groupby('age')['total_revenue'].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(age_revenue['age'], age_revenue['total_revenue'], marker='o')

plt.title('Rata-rata Revenue Berdasarkan Usia Pelanggan')
plt.xlabel('Usia')
plt.ylabel('Rata-rata Revenue')

plt.grid(True)
plt.show()
```

**Insight:**

Grafik menunjukkan bahwa rata-rata revenue pelanggan tersebar di berbagai usia tanpa pola peningkatan yang sangat tajam. Hal ini menandakan bahwa pelanggan dari berbagai kelompok usia tetap memberikan kontribusi pembelian.

**Action:**

Perusahaan dapat tetap mempertahankan strategi pemasaran yang menjangkau berbagai kelompok usia, namun tetap melakukan segmentasi agar promosi lebih tepat sasaran.

# **4. Histogram**

**Melihat distribusi total revenue pelanggan**

```python
plt.figure(figsize=(8,5))

sns.histplot(clustering_data['total_revenue'],
             bins=15,
             kde=True,
             color='skyblue')

plt.title('Distribusi Total Revenue Pelanggan')
plt.xlabel('Total Revenue')
plt.ylabel('Frekuensi')

plt.show()
```

**Insight:**

Histogram menunjukkan bahwa sebagian besar pelanggan memiliki total revenue pada nilai rendah hingga menengah, sedangkan hanya sedikit pelanggan yang memiliki revenue sangat tinggi.

**Action:**

Perusahaan dapat meningkatkan pembelian pelanggan melalui promo bundling atau diskon pembelian, serta mempertahankan pelanggan dengan revenue tinggi melalui program loyalitas.

# **5. BOXPLOT**

**Melihat penyebaran harga rata-rata produk**

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=clustering_data[['total_quantity','total_revenue','avg_price','age']])

plt.title('Distribusi Variabel Pelanggan')
plt.xlabel('Variabel')
plt.ylabel('Nilai')

plt.show()
```

**Insight:**

Boxplot menunjukkan bahwa total_revenue memiliki penyebaran nilai paling besar dibandingkan variabel lain. Hal ini menandakan adanya perbedaan cukup besar antara pelanggan dengan nilai pembelian rendah dan pelanggan dengan pembelian tinggi.

**Action:**

Perusahaan dapat mengidentifikasi pelanggan dengan revenue tinggi untuk diberikan program loyalitas atau promosi khusus agar mereka tetap aktif melakukan pembelian.

# **6. SCATTER PLOT**

**Melihat hubungan jumlah pembelian dan revenue**

```python
plt.figure(figsize=(8,5))

plt.scatter(clustering_data['total_quantity'],
            clustering_data['total_revenue'],
            alpha=0.3,
            s=20,
            color='teal')

plt.title('Hubungan Total Quantity dan Total Revenue')
plt.xlabel('Total Quantity')
plt.ylabel('Total Revenue')

plt.grid(True)
plt.show()
```

**Insight:**

Grafik menunjukkan bahwa semakin banyak produk yang dibeli pelanggan, maka total revenue yang dihasilkan juga cenderung meningkat.

**Action:**

Perusahaan dapat mendorong pelanggan membeli lebih banyak produk melalui promo bundling, paket produk, atau diskon pembelian dalam jumlah tertentu.

# **7. BUBBLE CHART**

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x='total_quantity',
    y='total_revenue',
    size='avg_price',
    hue='age',
    sizes=(20,400),
    alpha=0.4,
    data=clustering_data,
    palette='viridis'
)

plt.title('Hubungan Quantity dan Revenue dengan Variasi Harga')
plt.xlabel('Total Quantity')
plt.ylabel('Total Revenue')

plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```

**Insight:**

Bubble chart menunjukkan bahwa semakin besar jumlah pembelian pelanggan, maka total revenue juga cenderung meningkat. Ukuran bubble menggambarkan variasi harga rata-rata produk yang dibeli pelanggan.

**Action:**

Perusahaan dapat meningkatkan revenue dengan strategi yang mendorong pelanggan membeli lebih banyak produk atau memilih produk dengan harga lebih tinggi.

# **8. HEATMAP**

**Melihat korelasi antar variabel**

```python
plt.figure(figsize=(8,6))

sns.heatmap(clustering_data.corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title('Heatmap Korelasi Antar Variabel')

plt.show()
```

**Insight:**

Heatmap menunjukkan bahwa total_quantity memiliki hubungan korelasi cukup kuat dengan total_revenue, yang berarti semakin banyak produk yang dibeli pelanggan maka revenue juga meningkat.

Sementara itu variabel avg_price dan age tidak memiliki korelasi yang terlalu kuat terhadap variabel lainnya.

**Action:**

Perusahaan dapat meningkatkan revenue dengan mendorong peningkatan jumlah pembelian produk, misalnya melalui promo paket produk atau diskon pembelian dalam jumlah tertentu.
